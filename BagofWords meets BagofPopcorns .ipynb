```python
# Write all these steps in one single function

def review_to_words (raw_review):
    
    # Function to convert a raw review to a string of words
    # The input is a single string (a raw movie review), and 
    # the output is a single string (a preprocessed movie review)
    
    # Remove the HTML characters
    review_text = BeautifulSoup(raw_review)
    
    # Remove the punctuals
    letters_only = re.sub("[^a-zA-Z]"," ",review_text.get_text())
    
    # Convert it in lower case and split them into words
    lower_case = letters_only.lower()
    words = lower_case.split()
    
    #In Python, searching a set is much faster than searching
    #   a list, so convert the stop words to a set
    stops = set(stopwords.words("english")) 
    
    meaningfull_words = [w for w in words if not w in stops]
    
    
    #Join the words back into one string separated by space, 
    # and return the result.
    return( " ".join( meaningfull_words ))
    
```

```python
clean_review= review_to_words(train.review[0])
print clean_review
```

```python
num_reviews = train['review'].size
num_reviews
```

```python
clean_train_reviews= []

for i in xrange(0, num_reviews):
    # If the index is evenly divisible by 1000, print a message
    if( (i+1)%5000 == 0 ):
        print "Review %d of %d\n" % ( i+1, num_reviews ) 
    
    clean_train_reviews.append(review_to_words(train.review[i]))
```

## Playing with Stemming and Lemmatization

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()

clean_train_reviews_1= [[porter_stemmer.stem(word) for word in sentence.split(" ")] for sentence in clean_train_reviews]
```

```python
clean_train_reviews_1= [" ".join(sentence) for sentence in clean_train_reviews_1]
```

```python
clean_train_reviews_1
```

```python
print "creating the bag of words"

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = "word", tokenizer = None, preprocessor = None, stop_words = None,max_features = 5000)


# fit_transform() does two functions: First, it fits the model
# and learns the vocabulary; second, it transforms our training data
# into feature vectors. The input to fit_transform should be a list of 
# strings.

train_data_features = vectorizer.fit_transform(clean_train_reviews_1)

# Numpy arrays are easy to work with, so convert the result to an 
# array
train_data_features = train_data_features.toarray()

print (train_data_features).shape

print train_data_features
```

```python
vocab= vectorizer.get_feature_names()
print vocab
```

```python
train_data_features.shape
```

## Do the train_test_split to do cross-validation at this point


```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_data_features, train["sentiment"], test_size =0.2, random_state=0)
```

```python
print X_train.shape, y_train.shape
```

```python
# count of each distinct words in the vocab

dist = np.sum(train_data_features, axis=0)

# print the vocabulary word and count how many times it appears

for tag, value in zip(dist, vocab):
    print value, tag
```

## Supervised Models

```python
## Random Forest

from sklearn.ensemble import RandomForestClassifier
```

```python
start = time()
forest = RandomForestClassifier(n_estimators = 200, max_features= "log2", min_samples_split=200)

forest.fit(X_train, y_train)

end = time()

training_time= end - start

print ' Training Time:', training_time
```

```python
start= time()
pred_sentiment_train= forest.predict(X_train)
pred_sentiment_test = forest.predict(X_test)
end = time()

pred_time = end - start
print" prediction time" , pred_time
```

##ROC score

```python
from sklearn.metrics import roc_auc_score

print "train score: ", roc_auc_score(y_train,pred_sentiment_train)
print "test score: ", roc_auc_score(y_test,pred_sentiment_test)
```
